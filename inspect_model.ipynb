```python
"""
inspect_model.py
================
Run this ONCE locally (or add temporarily to your Render service) to print
the exact feature column names your final_model.pkl expects.

Usage:
    python inspect_model.py

Then paste the output here so we can fix predict.php to send the right names.
"""
import pickle

with open("final_model.pkl", "rb") as f:
    pkg = pickle.load(f)

print("=== Keys in pickle ===")
print(list(pkg.keys()))

print("\n=== feature_columns ===")
cols = pkg.get("feature_columns") or pkg.get("features") or pkg.get("columns")
if cols is not None:
    for i, c in enumerate(cols):
        print(f"  [{i}] {c}")
else:
    print("  Not found under 'feature_columns', 'features', or 'columns'")
    print("  All keys:", list(pkg.keys()))

print("\n=== model type ===")
model = pkg.get("model") or pkg.get("classifier") or pkg.get("clf")
print(" ", type(model))
if hasattr(model, "feature_names_in_"):
    print("  feature_names_in_:", list(model.feature_names_in_))
if hasattr(model, "n_features_in_"):
    print("  n_features_in_:", model.n_features_in_)

print("\n=== threshold ===")
print(" ", pkg.get("threshold", "not set (default 0.5)"))

print("\n=== scaler ===")
scaler = pkg.get("scaler")
print(" ", type(scaler))
if scaler and hasattr(scaler, "feature_names_in_"):
    print("  scaler feature_names_in_:", list(scaler.feature_names_in_))
```
